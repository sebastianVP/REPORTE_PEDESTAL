```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# Ruta al archivo HDF5
filename = "/content/2025-01-14T15-00-00/pos@1736869080.000.h5"

# Función para graficar arreglos según su forma
def plot_data(data, title="Data"):
    """
    Crea una visualización del arreglo de datos dependiendo de su dimensión.
    """
    if data.ndim == 1:  # Datos unidimensionales
        plt.figure(figsize=(8, 5))
        plt.plot(data, label=title)
        plt.title(f"Gráfico de Línea: {title}")
        plt.xlabel("Índice")
        plt.ylabel("Valor")
        plt.legend()
        plt.grid(True)
    elif data.ndim == 2:  # Datos bidimensionales
        plt.figure(figsize=(8, 6))
        plt.imshow(data, aspect='auto', cmap='viridis')
        plt.colorbar(label="Valor")
        plt.title(f"Matriz de Calor: {title}")
        plt.xlabel("Dimensión 1")
        plt.ylabel("Dimensión 2")
    else:
        print(f"No se puede graficar el arreglo {title} con dimensiones {data.shape}.")
    plt.show()

# Exploración y graficación del archivo HDF5
with h5py.File(filename, "r") as obj:
    for key in obj.keys():
        print(f"Grupo: {key}")
        for key2 in obj[key].keys():
            param = f"{key}/{key2}"
            print(f"Atributo: {key2}")
            try:
                data = np.array(obj[param])
                print(f"  - Forma del arreglo: {data.shape}")
                plot_data(data, title=param)
            except Exception as e:
                print(f"  - No se pudo cargar o graficar el atributo {key2}. Error: {e}")
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# Función para calcular el promedio de 'azi_speed' en un archivo HDF5
def calculate_avg_speed(filename):
    try:
        with h5py.File(filename, "r") as obj:
            for key in obj.keys():
                for key2 in obj[key].keys():
                    if key2 == "azi_speed":
                        param = f"{key}/{key2}"
                        try:
                            data = np.array(obj[param])
                            avg_speed = np.mean(data)
                            return avg_speed
                        except Exception as e:
                            print(f"Error al procesar '{key2}' en {filename}: {e}")
                            return None
        print(f"'azi_speed' no encontrado en el archivo {filename}.")
        return None
    except FileNotFoundError:
        print(f"El archivo {filename} no existe.")
        return None
    except Exception as e:
        print(f"Ocurrió un error al leer el archivo {filename}: {e}")
        return None

# Lista de archivos HDF5
import os
# Directorio base
base_path = "/content/2025-01-14T15-00-00"
base_path = "/content/2024-12-27T17-00-00"

# Listar archivos en el directorio
file_list = os.listdir(base_path)

# Crear rutas completas utilizando os.path.join
full_paths = [os.path.join(base_path, file) for file in file_list]


# Calcular los promedios de 'azi_speed' para cada archivo
average_speeds = []
valid_files = []

for filename in full_paths:
    avg_speed = calculate_avg_speed(filename)
    if avg_speed is not None:
        average_speeds.append(avg_speed)
        valid_files.append(filename)

# Verificar si hay datos para graficar
if average_speeds:
    # Graficar los valores promedio
    plt.figure(figsize=(24, 6))
    bars = plt.bar(valid_files, average_speeds, color='skyblue')
    # Añadir valores promedio encima de las barras
    for bar, avg in zip(bars, average_speeds):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
                 f"{avg:.4f}", ha="center", va="bottom", fontsize=10, color='black')
    plt.xlabel("Archivos HDF5")
    plt.ylabel("Velocidad Promedio (azi_speed)")
    plt.title("Velocidad Promedio de 'azi_speed' por Archivo")
    plt.xticks(rotation=45, ha="right", fontsize=9)
    plt.tight_layout()
    plt.show()
else:
    print("No se encontraron datos válidos para 'azi_speed'.")
```

```python
import matplotlib
from plotting_codes import sophy_cb_tables
for name, cb_table in sophy_cb_tables:
    ncmap = matplotlib.colors.ListedColormap(cb_table, name=name)
    matplotlib.pyplot.register_cmap(cmap=ncmap)

PARAM = {
    'S': {'var': 'power'                     ,'vmin': -45, 'vmax': -15 ,'cmap': 'jet'    ,'label': 'Power'         ,'unit': 'dBm'},
    'V': {'var': 'velocity'                  ,'vmin': -10, 'vmax':  10 ,'cmap': 'sophy_v','label': 'Velocity'      ,'unit': 'm/s'},
    'Z': {'var': 'reflectivity'              ,'vmin': -20, 'vmax':  80 ,'cmap': 'sophy_z','label': 'Reflectivity'  ,'unit': 'dBZ'},
    'W': {'var': 'spectral_width'            ,'vmin':  0 , 'vmax':  12 ,'cmap': 'sophy_w','label': 'Spectral Width','unit': 'm/s'},
    'R': {'var': 'rhoHV'                     ,'vmin': 0.2, 'vmax':  1  ,'cmap': 'sophy_r','label': 'RhoHV'         ,'unit': ' '  },
    'D': {'var': 'differential_reflectivity' ,'vmin': -9 , 'vmax':  12 ,'cmap': 'sophy_d','label': 'ZDR'           ,'unit': 'dB' }
    }

def read_atributos(filename,variable,channel,PARAM):
    with h5py.File(filename, "r") as obj:
        var        = PARAM[variable]['var']
        var_       = 'Data/'+var+'/'+str(channel)
        data_arr   = numpy.array(obj[var_])                 # data
        utc_time   = numpy.array(obj['Data/time'])
        data_azi   = numpy.array(obj['Metadata/azimuth'])   # th
        data_ele   = numpy.array(obj["Metadata/elevation"])
        heightList = numpy.array(obj["Metadata/range"])     # r
        unit       = PARAM[variable]['unit']
        cmap       = PARAM[variable]['cmap']
        vmin       = PARAM[variable]['vmin']
        vmax       = PARAM[variable]['vmax']
        label      = PARAM[variable]['label']
        return data_arr, utc_time,data_azi,data_ele,heightList,unit,cmap,vmin,vmax,label

def selectHeightsByIndex(heightList,minIndex, maxIndex):

    if (minIndex < 0) or (minIndex > maxIndex):
        raise ValueError("Height index range (%d,%d) is not valid" % (minIndex, maxIndex))

    if (maxIndex >= len(heightList)):
        maxIndex = len(heightList)

    new_h = heightList[minIndex:maxIndex]
    return new_h

def selectHeights(heightList,minHei,maxHei):

    if minHei and maxHei:
        if (minHei < heightList[0]):
            minHei = heightList[0]
        if (maxHei > heightList[-1]):
            maxHei = heightList[-1]
        minIndex = 0
        maxIndex = 0
        heights = heightList

        inda = numpy.where(heights >= minHei)
        indb = numpy.where(heights <= maxHei)

        try:
            minIndex = inda[0][0]
        except:
            minIndex = 0

        try:
            maxIndex = indb[0][-1]
        except:
            maxIndex = len(heights)

        new_heightList= selectHeightsByIndex(heightList=heightList,minIndex=minIndex, maxIndex=maxIndex)

    return new_heightList, minIndex,maxIndex

def plot_WR(count,z,r,az,rf,my_time,label,unit,cmap,vmin,vmax,mode,channel,grado,RMAX):
    if count ==1:
        plt.rcParams['axes.facecolor']='black'
        #cgax, pm = wrl.vis.plot_ppi(data=z,r=r,az=az,rf=1,ax=111,proj='cg',cmap=cmap,vmin=vmin,vmax=vmax)
        fig = plt.figure(figsize=(16, 8))
        cgax, pm= wrl.vis.plot_ppi(data=z,r=r,az=az,rf=1,ax=111,fig=fig,cmap=cmap,vmin=vmin,vmax=vmax)
        #caax     = cgax.parasites[0]
        title    = 'Sophy Plot'+"-"+label+"-"+ my_time+" "+str(mode)+"_"+str(channel)+"_"+str(grado)+"_"+" N° "+str(count)
        t = plt.title(title, fontsize=12,y=1.05)
        cbar     = plt.gcf().colorbar(pm,pad=0.075)
        cgax.grid()
        #plt.text(1.0, 1.05, 'azimuth', transform=caax.transAxes, va='bottom',ha='right')
        plt.text(1.0, 1.05, 'azimuth', va='bottom',ha='right')
        #plt.grid(True)
        cbar.set_label(label+'[' + unit + ']')
        #gh = cgax.get_grid_helper()
        site=(0,0,0)
        wrl.vis.plot_ppi_crosshair(
        site=site,
        ranges=[0, RMAX/6.0,RMAX*2/6.0,RMAX*3/6,RMAX*4/6,RMAX*5/6.0,RMAX],
        angles=[0, 90, 180, 270],
        line=dict(color="white"),
        circle={"edgecolor": "white"},)
    else:
        fig = plt.figure(figsize=(16, 8))
        cgax, pm=  wrl.vis.plot_ppi(data=z,r=r,az=az,rf=1,ax=111,fig=fig,cmap=cmap,vmin=vmin,vmax=vmax)
        #cgax, pm = wrl.vis.plot_ppi(data=z,r=r,az=az,rf=1,ax=111,cmap=cmap,vmin=vmin,vmax=vmax)
        #caax  = cgax.parasites[0]
        title    = 'Sophy Plot'+"-"+label+"-"+ str(my_time)+" "+str(mode)+"_"+str(channel)+"_"+str(grado)+"_"+" N° "+str(count)
        t = plt.title(title, fontsize=12,y=1.05)
        cbar = plt.gcf().colorbar(pm, pad=0.075)
        cgax.grid()
        #plt.text(1.0, 1.05, 'azimuth', transform=caax.transAxes, va='bottom',ha='right')
        plt.text(1.0, 1.05, 'azimuth', va='bottom',ha='right')
        cbar.set_label(label+'[' + unit + ']')
        site=(0,0,0)
        wrl.vis.plot_ppi_crosshair(
        site=site,
        ranges=[0, RMAX/6.0,RMAX*2/6.0,RMAX*3/6,RMAX*4/6,RMAX*5/6.0,RMAX],
        angles=[0, 90, 180, 270],
        line=dict(color="white"),
        circle={"edgecolor": "white"},)
        #plt.grid(True)
        #gh = cgax.get_grid_helper()
        print("plot",count)
# LECTURA Y PROCESAMIENTO
def read_pp(PATH,LISTA,RMAX,RMIN,variable,mode,channel,grado,PARAM):
    count     = 0
    len_Files = len(LISTA)
    for thisFile in LISTA:
        count += 1
        print("Count :",count)
        fullPathFile = PATH+ thisFile
        # LECTURA DE ARCHIVOS Y EXTRACCION DE ATRIBUTOS
        data_arr,utc_time,data_azi,data_ele,heightList,unit,cmap,vmin,vmax,label=read_atributos(filename=fullPathFile,variable=variable,channel=channel,PARAM=PARAM)
        # SELECCION DE ALTURAS
        if RMAX == 0:
           RMAX= heightList[-1]
        if RMIN == 0:
           RMIN = 0.01
        new_heightList,minIndex,maxIndex = selectHeights(heightList,RMIN,RMAX)
        # TIEMPO
        utc_time[0] = utc_time[0]+60*60*5
        my_time    = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(utc_time[0]))
        time_save  = time.strftime('%Y%m%d_%H%M%S',time.localtime(utc_time[0]))
        # VARIABLES
        len_X = data_arr.shape[0]
        x     = numpy.linspace(1,len_X,len_X)
        y     = new_heightList
        z_wrl = data_arr[:,minIndex:maxIndex]
        z     = data_arr[:,minIndex:maxIndex].transpose()
        if count==1:
          plot_WR(count=count,z=z_wrl,r=y,az=data_azi,rf=1,my_time=my_time,label=label,unit=unit,cmap=cmap,vmin=vmin,vmax=vmax,mode=mode,channel=channel,grado=grado,RMAX=RMAX)
        else:
          #pass
          plot_WR(count=count,z=z_wrl,r=y,az=data_azi,rf=1,my_time=my_time,label=label,unit=unit,cmap=cmap,vmin=vmin,vmax=vmax,mode=mode,channel=channel,grado=grado,RMAX=RMAX)
        plt.show()


#Convert toRAINFALL INTENSITY(unit: mm/h) using the Marshall -PALME Z(R) Parameters
def getRRfromZDB(Z,a,b):
    Z_lineal=  10**(Z/10.0) #(mm**6/mm**-3)
    R       = (Z_lineal/a)**(1./b)
    return R

#Convert to Rainfall Depth(Unit:mm)
def r_to_Depth(R,scan_time):
    #SCAN TIME IN SECONDS
    depth= R*scan_time/3600
    return depth
```
